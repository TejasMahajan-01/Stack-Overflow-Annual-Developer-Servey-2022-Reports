<h2> Let's Find Out The Top 10 Most Popular Programming Languages Of The Year 2022. </h2>

It appears that a most of respondents have selected more than one language among the options. We need to split them into single data value.

````python
### Defining a function to split data into single data value in each row

def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in col_series[col_series.notnull()].iteritems():
        # Break each value into list of options
        for option in value.split(';'):
            # Add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Mark the value in the option column as True
            result_df.at[idx, option] = True
    return result_df[options]
````

_______

````python
popular_lang = split_multicolumn(df1.LanguageHaveWorkedWith) # Splitting data into single values
popular_lang
````

Let's aggregate these to identify the percentage of respondents who selected each language.

````python
# Percentage of respondents who selected each language.
popular_language_percentage = popular_lang.mean().sort_values(ascending=False)*100
df_popular_language_percentage = popular_language_percentage.to_frame().reset_index().round(2)
df_popular_language_percentage.columns = ['PopularLanguages', 'Percentage']
df_popular_language_percentage
````

Do you want to know how the `Mean` function returns a percentage from True and False values, then check out this explanation.

<details><summary> Explanation</summary>
    
    
```python
    #The Mean fuctions performs these operation:
    Sum = popular_lang['JavaScript'].sum()       # Sum is the total of true values
    Count = popular_lang['JavaScript'].count()   # Counts is the number of rows
    Percentage = round((Sum/Count*100), 2)      # Rounding values to two decimal places
    print('Percentage of JavaScript:', Percentage) # It will returns the percentage for JavaScript which 64.71%
    #The mean function performs these operations for each column.
```

</details>

````python
# Creating percentage values for plotting value labels in each bar of a bar plot
df_popL_percentage = df_popular_language_percentage['Percentage'].astype(str) + '%'
df_popL_percentage
list_popL_percentage = df_popL_percentage.head(10).to_list()
print(list_popL_percentage)
````

````python
# Visualizing the distribution of data with Seaborn package
ax = sns.barplot(y = 'PopularLanguages', x = 'Percentage', data =df_popular_language_percentage.head(10), orient="h")
ax.set_title('Top 10 Most Popular Programming Languages In Year 2022', fontdict={'size':20}, y=1.05)
ax.set_xlabel('Percentage of Respondents', fontdict={'size':17}, labelpad=25)
ax.set_ylabel('Languages', fontdict={'size':17}, labelpad=25)
ax.bar_label(ax.containers[0],labels=list_popL_percentage, padding=-50)
sns.set(rc={'figure.figsize':(15,8)})
plt.show()
````

<h3> Insights and Inferences </h3>

* JavaScript is the most popular and commonly used programming language of the year 2022, with around 64.71% of the respondents choosing to work with JavaScript.
* HTML/CSS, SQL and Python are also the top and the most commonly used programming languages in the year 2022.

<h2> Let's Find Out Which Age Group Of The Survey Respondents Are Getting Higher Salary. </h2>

````python
df_fltr = df1.drop(labels=['ResponseId', 'ConvertedCompYearly'], axis = 1) # Droping unwanted column for calculating median of dataframe
````

````python
df_comp = df_fltr.groupby(['Age']).median().sort_values(by = 'CompTotal', ascending = False).rename_axis('AgeGroup').reset_index()
df_comp
````

````python
# Visualizing the distribution of data with Seaborn package
ax = sns.barplot(x='CompTotal', y='AgeGroup', data=df_comp, orient='h' )
ax.set_title('Median Salary Of Different Age Groups Of The Survey Respondents', y = 1.05, fontdict = {'size':20})
ax.set_xlabel('Median Salary In USD', fontdict={'size':17}, labelpad=25)
ax.set_ylabel('Age Group', fontdict={'size':17}, labelpad=25)
ax.bar_label(ax.containers[0], padding=-50)
sns.set(rc={'figure.figsize':(15,8)})
plt.show()
````

<h3> Insights and Inferences </h3>

* Respondents who are in the age group of 55-64 years and respondents who choose not to say during the survey are getting higher salary, followed by 65 years or above age group and 45-54 years age group.
